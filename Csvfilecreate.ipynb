```python
companies = ['1Huddle', 'Oto Capital', '3D Look','3desk','3TEN8','4Vets','8sec','9FIN']
industries = ['Consumer', 'Financial Technology', 'Consumer','Software and Services','Software and Services','Consumer','Consumer','Financial Technology']
sub_industries = ['Gaming & eSports Job & Career Service', 'Credit & Lending','Fashion' ,'Human Resources & Recruiting','Developer Tools','Animals','Gaming & eSports','Asset Management']
countries = ['United States', 'India', 'United States','United Kingdom','United States','Brazil','France','United Kingdom']
```

```python
data = zip(companies, industries, sub_industries, countries)
```

```python
csv_file = 'data.csv'
```

```python
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the header row
    writer.writerow(['Company Name', 'Industry', 'Sub-Industry', 'Country'])
    
    # Write the data rows
    for row in data:
        writer.writerow(row)

print(f'CSV file "{csv_file}" has been created.')
```
